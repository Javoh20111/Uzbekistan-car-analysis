```python
def foreign_key_maker(df, column, id_make):
    lookup = df[[column]].drop_duplicates().reset_index(drop=True)
    lookup[id_make] = lookup.index + 1

    df = df.merge(lookup, on=column, how='left')

    return df, lookup

df, brand = foreign_key_maker(df, 'brand', 'brand_id')
df, currency = foreign_key_maker(df, 'currency', 'currency_id')
```

```python
def make_str(df, column):
    df[column] = df[column].str.split(',')
    
    df[column] = df[column].apply(
        lambda x: [i.strip() for i in x] if isinstance(x, list) else x
    )
    
    return df
make_str(df, 'sale_type')
make_str(df, 'additional_options')
```

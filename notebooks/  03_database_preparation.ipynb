```python
def foreign_key_maker(df, column, id_make):
    lookup = df[[column]].drop_duplicates().reset_index(drop=True)
    lookup[id_make] = lookup.index + 1

    df = df.merge(lookup, on=column, how='left')

    return df, lookup

df, brand = foreign_key_maker(df, 'brand', 'brand_id')
df, currency = foreign_key_maker(df, 'currency', 'currency_id')
df, condition = foreign_key_maker(df, 'condition', 'condition_id')
df, color = foreign_key_maker(df, 'color', 'color_id')
df, transmission = foreign_key_maker(df, 'transmission', 'transmission_id')
df, fuel_type = foreign_key_maker(df, 'fuel_type', 'fuel_type_id')
df, body_type = foreign_key_maker(df, 'body_type', 'body_type_id')
df, region = foreign_key_maker(df, 'region', 'region_id')
df, district = foreign_key_maker(df, 'district', 'district_id')
```

```python
df.head()
```

```python
def make_str(df, column):
    df[column] = df[column].str.split(',')
    
    df[column] = df[column].apply(
        lambda x: [i.strip() for i in x] if isinstance(x, list) else x
    )
    
    return df
make_str(df, 'sale_type')
make_str(df, 'additional_options')
```

```python
from sqlalchemy import create_engine

engine = create_engine('postgresql://postgres:8228@localhost:5432/cars_db')



def inserter(filepath, table_name):
    df = pd.read_csv(filepath)

    df.to_sql(
        table_name,
        engine,
        if_exists='append',
        index=False
    )

    print(f'Data inserted into {table_name}')
```

```python
transmissions = transmission[['transmission_id', 'transmission']].rename(columns={'transmission': 'transmission_name'})
transmissions.to_csv('../data/Prepared/transmissions.csv', index=False)

colors = color[['color_id', 'color']].rename(columns={'color':'color_name'})
colors.to_csv('../data/Prepared/colors.csv', index=False)

conditions = condition[['condition_id', 'condition']].rename(columns={'condition':'condition_name'})
conditions.to_csv('../data/Prepared/conditions.csv',index=False)

districts = district[['district_id', 'district']].rename(columns={'district': 'district_name'})
districts.to_csv('../data/Prepared/districts.csv',index=False)

fuel_types = fuel_type[['fuel_type_id', 'fuel_type']].rename(columns={'fuel_type':'fuel_type_name'})
fuel_types.to_csv('../data/Prepared/fuel_types.csv',index=False)

regions = region[['region_id', 'region']].rename(columns={'region':'region_name'})
regions.to_csv('../data/Prepared/regions.csv',index=False)

currencies = currency[['currency_id', 'currency']].rename(columns={'currency':'currency_name'})
currencies.to_csv('../data/Prepared/currencies.csv',index=False)

inserter('../data/Prepared/brands.csv', 'brands')
inserter('../data/Prepared/conditions.csv', 'conditions')
inserter('../data/Prepared/colors.csv', 'colors')
inserter('../data/Prepared/transmissions.csv', 'transmissions')
inserter('../data/Prepared/fuel_types.csv','fuel_types')
inserter('../data/Prepared/regions.csv', 'regions')
inserter('../data/Prepared/districts.csv', 'districts')
inserter('../data/Prepared/sale_options.csv', 'sale_options')
inserter('../data/Prepared/additional_options.csv', 'additional_options')
inserter('../data/Prepared/cars.csv', 'cars')
```
